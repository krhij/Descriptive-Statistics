```python
import pandas as pd
from dfply import *
```

```python
import seaborn as sn
```

```python
cars = sn.load_dataset('mpg')
```

# Zad 1

```python
cars >> select(X.mpg, X.horsepower) >> head(3)
```

# Zad 3

```python
cars >> select(~X.model_year, ~X.name) >> head(3)
```

# Zad 5

```python
cars >> select(contains("a")) >> head(3)
```

# Zad 7

```python
cars >> row_slice([1,2,3])
```

# Zad 9

```python
cars >> mask(X.mpg > 20, X.origin == "japan") >> head(3)
```

# Zad 11

```python
cars >> mutate(mpg_per_cylinder = X.mpg/X.cylinders) >> head(3)
```

# Zad 13

```python
cars >> group_by(X.origin) >> mutate(mpg_lead = lead(X.mpg))
```

# Zad 15

```python
cars >> rename(miles_per_gallon = X.mpg)
```
